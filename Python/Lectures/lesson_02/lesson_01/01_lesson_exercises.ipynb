Task nr 1

Write code that:
- Allows the user to input numbers a and b (int or float)
- Prints "a is less than b" if that is true
- Prints "a is equal to b" if that is true
- Prints "a is greater than b" if that is true

Tip: Use input, if, elif, and else statements

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
if a < b:
    print(f"{a} is less than {b}")
elif a > b:
    print(f"{a} is greater than {b}")
else:
    print(f"{a} is equal to {b}")
```

Task nr 2

Write code that performs the following operations on the string "Zen of Python":
- Prints the last character of the second word
- Prints the first character of the third word
- Prints only the first word
- Prints only the last word
- Prints the entire phrase in reverse
- Splits the words and prints them separately
- Replaces the word "Python" with "Programming" and prints the new sentence

Tip: Use string slicing, and the functions split() and replace().
The Zen of Python:

```python
zen_of_python = "Zen of Python"
print(zen_of_python.split()[1][-1])
print(zen_of_python.split()[2][0])
print(zen_of_python.split()[0])
print(zen_of_python.split()[-1])
print(zen_of_python[::-1])
print(zen_of_python.split())
print(zen_of_python.replace("Python", "Programming"))
```

Task nr 3

In the program, try out more string functions:
- upper()
- casefold()
- capitalize()
- count()
- find()
- etc.

```python
zen_of_python = "Zen of Python"
print(zen_of_python.upper())
print(zen_of_python.casefold()) #lowercase
print(zen_of_python.capitalize()) # first letter uppercase
print(zen_of_python.count("o")) # count how many "o" in the string
print(zen_of_python.find("P")) # count the index of the first "P" in the string
```

Task nr 4

Write a program that:
- Allows the user to enter the first number
- Allows the user to enter the second number
- Asks which mathematical operation should be performed
- Prints the result: the sum, product, or other operation of the chosen numbers

Tip: Use input(), if, and print()

```python
x = int(input("First number: "))
operation = input("Enter mathematical operation (+, -, *, /): ")
y = int(input("Second number: "))
if operation == "+":
    Result = x + y
elif operation == "-":
    Resukt: x - y
elif operation == "*":
    Result: x * y
elif operation == "/":
    Result: x / y
else:
    Result = "INVALID INPUT"
print(f"Result is: {Result}")
```
