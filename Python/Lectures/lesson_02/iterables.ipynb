```python
for num in list1:
    if num % 2 == 0:
        print(num)
```

```python
for num in list1:
    if num % 2 == 0:
        print(num)
    else:
        print('Odd number')
```

```python
# Start sum at zero
list_sum = 0
for num in list1:
    list_sum = list_sum + num
print(list_sum)
```
