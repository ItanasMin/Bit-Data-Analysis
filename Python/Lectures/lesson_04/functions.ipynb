```python
def line_total(qty, unit_price, discount=0.0):
    return round(qty * unit_price * (1 - discount), 2)
 
print(line_total(2, 1.50, 0.00))
print(line_total(1, 0.99, 0.10))
```

```python
def add_vat(amount, vat_rate=0.21):
    return round(amount * (1 + vat_rate), 2)
 
print(add_vat(100))
print(add_vat(100, vat_rate=0.09))
```

```python
def safe_divide(a, b):
    if b == 0:
        return None
    return a / b
 
print(safe_divide(10, 2))
print(safe_divide(10, 0))
```

```python
def customer_segment(total_spent):
    if total_spent >= 200:
        return "Aukštos vertės"
    elif total_spent >= 100:
        return "Vidutinės vertės"
    elif total_spent > 0:
        return "Žemos vertės"
    else:
        return "Neteisinga suma"
 
print(customer_segment(250))
print(customer_segment(120))
print(customer_segment(20))
print(customer_segment(0))
```

```python
def rodyti_info(**kwargs):
    for raktas, reiksme in kwargs.items():
        print(f"{raktas}: {reiksme}")
 
rodyti_info(vardas="Jonas", amzius=30, miestas="Panevėžys")
```

```python
def visa_info(vardas, *hobiai, **kontaktai):
    print(f"Vardas: {vardas}")
    print("Hobiai:", hobiai)
    print("Kontaktai:")
    for raktas, reiksme in kontaktai.items():
        print(f"  {raktas}: {reiksme}")
 
visa_info("Jonė", "dailė", "joga", "katės", "bėgimas", "kelionės", elpastas="jone45893@gmail.com", miestas="Jonava")
```

```python
balansas = 1000
 
def papildyti_balansa(suma):
    global balansas
    balansas += suma
    print("Atnaujintas balansas:", balansas)
 
papildyti_balansa(500)
print("Galutinis balansas:", balansas)
```

```python
def papildyti_balansa_teisingai(balansas, suma):
    return balansas + suma
 
balansas = 1000
balansas = papildyti_balansa_teisingai(balansas, 300)
print("Galutinis balansas:", balansas)
```

```python
def skaiciuoti_konversija(parodymai, paspaudimai):
    konversija = round((paspaudimai / parodymai) * 100,2)
    return konversija
 
rezultatas = skaiciuoti_konversija(10000, 1450)
print("Konversija: ", rezultatas, "%")
```
