8. Apskaičiuokite, kiek procentų liko negrąžintos paskolos (1000 € – 350 € grąžinta).

```python
visa_paskola = 1000
grazinta = 350
likutis = visa_paskola - grazinta
likutis_procentais = (likutis / visa_paskola) * 100
print(likutis_procentais)
```

21. Patikrinkite, ar skaičius yra teigiamas, neigiamas ar nulis.

```python
skaicius = int(input("Įveskite skaičių: "))
if skaicius > 0:
    print("skaicius teigiamas")
elif skaicius < 0:
    print("skaicius neigiamas")
else:
    print("nulis")
```

24. Įveskite vardą – jei „admin“, pasisveikinkite kaip su administratoriumi.
