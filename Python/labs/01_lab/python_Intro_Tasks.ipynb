21. Patikrinkite, ar skaičius yra teigiamas, neigiamas ar nulis.

```python
number = int(input("Enter a number: "))
if number > 0:
    print("The number is positive")
elif number < 0:
    print("The number is negative")
else:
    print("Zero")
```

23. Patikrinkite, ar skaičius yra lyginis ar nelyginis.

24. Įveskite vardą – jei „admin“, pasisveikinkite kaip su administratoriumi. 

```python
name = input("Enter your name here: ").lower().strip()
if name == "admin":
    print("Hello, Administrator")
else:
    print(f"Hello, {name}")
```

25. Įveskite temperatūrą: jei >25 – „karšta“, tarp 10–25 – „normali“, kitaip – „šalta“.

```python
temperature = float(input("Enter the temperature here: "))
if temperature > 25:
    print("hot")
elif temperature >= 10:
    print("normal")
else:
    print("cold")
```
